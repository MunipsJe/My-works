```python
## Filter the dataset for the Differnt City tier & Hospital Tier
tier1=df[df['City tier']=='tier - 1']
tier2=df[df['City tier']=='tier - 2']
tier3=df[df['City tier']=='tier - 3']
htier1=df[df['Hospital tier']=='tier - 1']
htier2=df[df['Hospital tier']=='tier - 2']
htier3=df[df['Hospital tier']=='tier - 3']
```

```python
## Extract the chrges columnfor each tiers:
tier1_chrge=tier1['charges']
tier2_chrge=tier2['charges']
tier3_chrge=tier3['charges']
htier1_chrge=htier1['charges']
htier2_chrge=htier2['charges']
htier3_chrge=htier3['charges']
```

```python
#Filter the dataset for smoker and non smoker
Smoker_yes=df[df['smoker']=='yes']
Smoker_no=df[df['smoker']=='No']
```

```python
## Extract the chrges column for smoker and non smoker
Sy_chrge=Smoker_yes['charges']
Sn_chrge=Smoker_no['charges']
```

```python
data=pd.crosstab(df['smoker'],df['Heart Issues'])
data
```

```python
#observed values
ob_val=data.values
ob_val
```

```python
#row total
row_total=ob_val.sum(axis=1)
#Column total
col_total=ob_val.sum(axis=0)
#Grand total
grd_total=ob_val.sum()
```

```python
#Finding Expected values
ex_val=np.outer(row_total,col_total)/grd_total
ex_val
```

```python
#Coputation of chi^2 value
chi_sq=(((ob_val-ex_val)**2)/ex_val).sum()
chi_sq
```

```python
#degree of freedom
ddf=(ob_val.shape[0]-1)*(ex_val.shape[1]-1)
ddf
```

```python
#Computionof p_value-Smoking and heart deisease are independent
p_value=1-stats.chi2.cdf(chi_sq,ddf)
p_value
```

The P-value was greater than >0.05 here we failed to reject Null Hypothesis , Hence smoking  and heart disease are independent

```python
#Cost between City  Tiers 1 & 2 
t_stat, p_value = stats.ttest_ind(tier1_chrge,tier2_chrge)
p_value
```

```python
#Cost between City  Tiers 1 & 3
t_stat, p_value = stats.ttest_ind(tier1_chrge,tier3_chrge)
p_value
```

```python
#Cost between City  Tiers 3 & 2 
t_stat, p_value = stats.ttest_ind(tier3_chrge,tier2_chrge)
p_value
```

The P-value was greater than >0.05 here we failed to reject Null Hypothesis , So there is no significance in  Charges between the Different tiers of City are not related

```python
#Cost between Hospital  Tiers 1 & 2
t_stat, p_value = stats.ttest_ind(htier1_chrge,htier2_chrge)
p_value
```

```python
#Cost between Hospital  Tiers 1 & 3
t_stat, p_value = stats.ttest_ind(htier1_chrge,htier3_chrge)
p_value
```

```python
#Cost between Hospital  Tiers 2 & 3
t_stat, p_value = stats.ttest_ind(htier3_chrge,htier2_chrge)
p_value
```

The P-value was less than >0.05 here  reject the Null Hypothesis , So there is significance in  Charges between the Hospital tiers .

```python
# Average charge distrobution between a smoker and non smoker
t_stat, p_value=stats.ttest_ind(Sy_chrge,Sn_chrge)
p_value
```

The P-value was less than >0.05 here  reject the Null Hypothesis , So there is significance in  Charges between a smoker and a non smoker is evident.
